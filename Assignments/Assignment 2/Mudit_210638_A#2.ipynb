What is the mean, median and standard deviation of 'Margin' in the 
dataset ?

```python
mean = df['Margin'].mean()
median = df['Margin'].median()
mode = df['Margin'].mode()
stddev = df['Margin'].std()
print('Mean =', mean, '\nMedian =', median, '\nMode =', mode[0], '\nStandard Deviation =', stddev)
```

Which player has received the most number of Player of the match award?

```python
most_potm = df['Player_of_Match'].mode()
print(most_potm[0])
```

Name the Team which has the highest Margin of Win ?

```python
margin_arr = df['Margin'].fillna(0)
index = margin_arr.argmax()
team_margin = df['WinningTeam'].to_numpy()
print(team_margin[index])
```

Make a pie chart to analyse Teams (Either Team 1 or Team 2) with number of match played?
Label the Pie Chart.

```python
arr = df['Team1'].to_numpy()
uniq_teams, num_matches = np38.unique(arr, return_counts = True)
plt38.figure(figsize = (10,10))
plt38.pie(num_matches, labels = uniq_teams)
plt38.show()
```

Bar chart of Umpires (U can use Umpire 1 or Umpire 2, ur choice) and number of match they umpired? 
Set figure size as (100,20)

```python
ump_arr = df['Umpire2'].to_numpy()
uniq_ump, matches_ump = np38.unique(ump_arr, return_counts = True)
plt38.figure(figsize = (100,50))
plt38.bar(uniq_ump, matches_ump)
plt38.xticks(rotation = 90, fontsize = 75)
plt38.yticks(fontsize = 100)
plt38.show()
```
